**Zadanie 4**

Stwórz tablicę 4x4, która zawiera wartości 1,2,3,4 w każdym wierszu.
Następnie wykorzystując funkcje łączenia tablic utwórz tablicę połączoną z powyższej tablicy i jej wersji lustrzanej w poziomie (spróbuj użyć wycinków). Postać każdego wiersza po połączeniu: `[1 2 3 4 4 3 2 1]`


```python
num = [1,2,3,4]
F = np.array([num for _ in range(4)])
print(F, end="\n\n")
#fun. lączenia tablic
#A to lustarna wersja F

A = F[::-1,::-1]
print(A, end="\n\n")

result = np.concatenate((F,A), axis=1)
print(result, end="\n\n")

```

**Zadanie 5**

Utwórz tablicę o nazwie `Z_5` o wymiarach 2x2 wypełnioną wartościami 1.
Wykorzystując funkcje łączenia tabel uwtórz tablicę, która finalnie będzie tablicą składającą się z 4-ech tablic `Z_5`, ale tak aby finalnie była to tablica 4 x 4 podobna do poniższej:

1,1,2,2  
1,1,2,2  
3,3,4,4  
3,3,4,4  

```python
Z_5 = np.ones((2,2), dtype=np.int32)
print(Z_5, end="\n\n")

# Tworzenie 4 bloków
Z_4 = Z_5[[0,1]]
Z_4[:] = 2
print(Z_4, end="\n\n")

Z_3 = Z_5[[0,1]]
Z_3[:] = 3
print(Z_3, end="\n\n")

Z_2 = Z_5[[0,1]]
Z_2[:] = 4
print(Z_2, end="\n\n")

part1 = np.concatenate((Z_5,Z_4), axis=1)
print(part1, end="\n\n")

part2 = np.concatenate((Z_3,Z_2), axis=1)
print(part1, end="\n\n")

final = np.concatenate((part1,part2), axis=0)
print(final, end="\n\n")
```

```python

Z_5 = np.ones((2, 2), dtype=int)

# Tworzenie 4 bloków
A = Z_5 * 1
B = Z_5 * 2
C = Z_5 * 3
D = Z_5 * 4

part1 = np.hstack((A,B))
print(part1, end="\n\n")

part2 = np.hstack((C,D))
print(part2, end="\n\n")

final = np.vstack((part1,part2))
print(final)
```
