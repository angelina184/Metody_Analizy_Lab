Zadanie 5
Bazując na przykładach powyżej oraz wykresie z punktu 2.2 wyświetl tę samą tablicę (3 x 3) w poniższych postaciach:

- widoczne są tylko elementy dla współrzędnych y = 1 
- widoczne są wartości dla z = 1 oraz z = 3 
- wszystkich wartości x oraz y (to tak jakby dwie kolumny tej kostki 3 x 3 x 3 - pierwsza i ostatnia).

```python
import numpy as np
import matplotlib.pyplot as plt

#Zadanie było przedstawione w sposób trochę niejasny i trudny do zrozumienia.

# współrzędne (x - wiersz, y - kolumna, z - głębia)
arr_3d = np.zeros((3,3,3), dtype='i')
print(arr_3d)

```

```python
#wartości dla z = 1 i z = 3 gdy z = 3 to index jest out of range
arr_3d[:,:,0] = 1
arr_3d[:,:,2] = 1
arr_3d

# kolory – cała kostka zielona
colors = np.full(arr_3d.shape, 'orange', dtype=object)

# rysowanie 3D
ax = plt.figure().add_subplot(projection='3d')

ax.axes.set_xlim3d(left=0, right=5) 
ax.axes.set_ylim3d(bottom=0, top=5) 
ax.axes.set_zlim3d(bottom=0, top=5) 

ax.voxels(arr_3d, facecolors=colors, edgecolor='k')

plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z (głębia)')

plt.show()
```

```python
#widoczne są tylko elementy dla współrzędnych y = 1   yэ(0,1) --> (0,1,0)(1,1,0)(2,1,0)
arr_3d[:,1,:] = 1
arr_3d

# kolory – cała kostka zielona
colors = np.full(arr_3d.shape, 'pink', dtype=object)

# rysowanie 3D
ax = plt.figure().add_subplot(projection='3d')

ax.axes.set_xlim3d(left=0, right=5) 
ax.axes.set_ylim3d(bottom=0, top=5) 
ax.axes.set_zlim3d(bottom=0, top=5) 

ax.voxels(arr_3d, facecolors=colors, edgecolor='k')

plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z (głębia)')

plt.show()



```

```python
#wszystkich wartości x oraz y
arr_3d[:, 1 , 0] = 1
arr_3d[:, 1 , 2] = 1

# kolory – cała kostka zielona
colors = np.full(arr_3d.shape, 'blue', dtype=object)

# rysowanie 3D
ax = plt.figure().add_subplot(projection='3d')

ax.axes.set_xlim3d(left=0, right=5) 
ax.axes.set_ylim3d(bottom=0, top=5) 
ax.axes.set_zlim3d(bottom=0, top=5) 

ax.voxels(arr_3d, facecolors=colors, edgecolor='k')

plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z (głębia)')

plt.show()

```

Bazując na przykładzie z podpunktu 2.3.2 przygotuj test czasu wykonania dla trzech poniższych przypadków:

- inicjalizacja listy 10000 wartości całkowitoliczbowych 
-  przemnożenie każdego elementu przez liczbę 2 wykorzystując zapis z pętlą for
tak jak powyżej, ale z wykorzystaniem list comprehension (dla przypomnienia [wyrażenie for elem in kolekcja])
dla tablicy numpy

```python
from timeit import timeit

setup = """
import numpy as np
list10_000 = list(range(10_000))
"""

stmt1 = """list(range(10_000))"""  #bez spacji
stmt2 = """[num*2 for num in list10_000]"""
stmt3 = """arr: np.ndarray = np.array([num*2 for num in list(range(10_000))])"""

print(timeit(stmt=stmt1, number=1))
print(timeit(stmt=stmt2,setup=setup, number=1))
print(timeit(stmt=stmt3, setup= setup, number=1))
```
